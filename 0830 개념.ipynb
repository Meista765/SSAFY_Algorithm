# 비트 연산
### 비트와 바이트
- 1bit : 0과 1을 표현하는 정보의 단위
- 1Byte : 8bit 를 묶어 1Byte라고 한다.

### 비트 연산
- 컴퓨터의 CPU는 0과 1로 다루어 동작되며, 내부적으로 비트 연산을 사용하여 +,-,X 등을 계산한다.

### AND와 OR 비트연산자 이해하기
- a AND b : a,b 둘 다 1 일때만 결과가 1이다. 그 외에는 0
- a OR b : a,b 둘 중 하나만 1이면 결과가 1이다. 그 외에는 0
- & 연산자 : 비트단위로 AND 연산을 한다. 
- | 연산자 : 비트
- ex ) 1001 & 1101 = 1001 : 자리수끼리 비교
- ex ) 3&4 = 100 & 011 = 000
- ex ) 7&5 = 111&101 = 101
- ex ) 7|5 = 111|101 = 111


### 파이썬에서 2진수, 16진수, 10진수 변환하여 출력하기
- 2진수는 숫자 0과 소문자 b → 0b를 접두사로 붙여 표현한다 ex) 0b100 = 4
- 16진수는 숫자 0과 소문자 x → 0x를 접두사로 붙여 표현한다 ex) 0x10 = 16
- 10진수를 2와 16진수로 변환
  - `print(bin(10))` # 0b1010
  - `print(hex(10))` # 0xa
- 16진수와 2진수를 10진수로 변환
  - `print(int('1011',2))`  #11
  - `print(int('b',16))`    #11
- 예제문제
  - 0b11011110&0b11011 = 11010 = 16+8+2 = 26
    - 자리수 안맞는건 0으로 채워넣음
    - `print(bin(0b11011110&0b11011))`#0b11010
  - 0x4A3|25(10) = 0100/1010/0011|11001 = 11011 = 1211
    - `print(bin(0x4A3), bin(25))` #0b10010100011 0b11001
    - `print(bin(0x4A3|25))` # 0b10010111011
    - `print(0x4A3|25)` #1211

### XOR 과 NOT 연산자
- ^ : XOR(엑스오어) 연산자, OR처럼 동작되는데 둘 다 1인 경우는 0이다.
- 비트단위로 XOR 연산(같으면 0, 다르면 1) ex)num1 ^ num2
- ex ) 1011 ^ 1101 = 0110
- `a = bin(0b1011 ^ 0b1101)` #0b110
- 어떤 값이던 특정 수로 2회 XOR를 하면 원래 수로 돌아온다
  - ex ) 1011 ^ 0100 = 1111 → 1111^0100 = 1011
- NOT : ~A = 0101 : A가 아니다 → 0과1을 바꿈 → 1010

### 비트 연산자
- Left Shift << : 특정 수 만큼 비트를 왼쪽으로 밀어낸다
  - num <<2 : 왼쪽으로 2회 이동
  - `print(bin(0b1101 <<2))`  #0b110100
- Right Shift >> : 특정 수 만큼 비트를 오른쪽으로 밀어낸다(우측 비트들이 제거된다.)
  - num >>2 : 오른쪽으로 2회 이동
  - `print(bin(0b1101 >>2))` #0b11
  - ex) `print(7>>2, bin(7>>2))` # 1 0b1


```python
arr = [1,2,3,4]

# 각 자리를 쓴다/안쓴다 → 나올 수 있는 경우의 수 2가지
# 각 자리마다 2가지 경우의 수 → N 길이라면 → 2^N만큼의 경우의 수가 나올 수 있다

# 부분집합의 수
print(1<<len(arr))
for i in range(1 << len(arr)):
    print(i, end=' ') 
print()
```

```python
# i & (1<<N) : N 번째 비트가 0인지 아닌지 알 수 있다


# i의미 : i번째 부분집합
for i in range(1 << len(arr)) :
    for idx in range(len(arr)) :
        #i & (1<<idx)
        # -i 번째 부분집합에 idx 요소가 포함되어 있나요?
        if i & (1 <<idx) :
            print(arr[idx], end = ' ')
    print()
```

### 음수 표현 방법
- 컴퓨터는 음수를 "2의 보수"로 관리한다.
- 값을 반전한 이후에 +1을 한 것
- Ex) 0001 (1) → 1111 (-1) → 0001 + 1111 = 0
- 맨 앞자리 bit(MSB)는 음수 or 양수를 구분하는 비트이다.
  - 음수 앞자리는 올림비트 때문에 1이 붙음(보수취할때는 올림 비트 버려짐 : 빼고 출력함)
  - MSB(​​Most Significant Bit) 최상위 비트
- 참고 ) 컴퓨터가 2의 보수를 사용하여 음수를 관리하는 이유 : 뺄셈의 연산 속도를 올릴 수 있으며, +0과 -0을 따로 취급하지 않기 위해 사용한다.
- 2의 보수 예시
  - 10001의 2의 보수 → 수를 모두 뒤집고 +1을 한다 → 01110 +1 = 01111
  - 1111000의 2의 보수 → 수를 뒤집으면  0000111 이고 +1을 한다. → 0000111 + 1 = 0001000
- 2의 보수를 취한 수를 한번 더 2의 보수를 취하면 원래의 값으로 돌아온다


### NOT 연산자
- (~)NOT 연산자 : 모든 비트를 반전시킨다.
- 만약 8bit 일 때 ~(0001 1111) 이라면 값은 1110 0000이 된다.

#### 파이썬에서 활용하기
- `~4`를 시행하면 -5가 출력된다.
-`print(bin(4))`  # 0b100
- `print(bin(-4))` #-0b101
- `print(~4)`     #-5

```python
# 홀/짝 판단

num1, num2 = 10, 5
if num1 & 1 : #num % 2 == 1
    print('홀수')
else :
    print('짝수')
    
```

```python
# 2. 나누기
l, r = 0, 100
# 100 : 4
mid = (l+r) >> 1
# mid = 10 : 2
print(mid)
```
